```python
df = pd.read_excel("Mall_Customers.xlsx")

# Display basic info and first few rows
df.info()
df.head()
```

# **DATA** **PRE**-**PROCESSING**

```python
# Drop CustomerID since it's not useful for clustering
df.drop(columns=['CustomerID'], inplace=True)

# Encode 'Genre' (Male/Female) using Label Encoding
label_encoder = LabelEncoder()
df['Genre'] = label_encoder.fit_transform(df['Genre'])

# Standardize numerical features
scaler = StandardScaler()
df[['Age', 'Annual_Income_(k$)', 'Spending_Score']] = scaler.fit_transform(
    df[['Age', 'Annual_Income_(k$)', 'Spending_Score']]
)
```

# **Determine** **optimal** **number** **of** **clusters**

```python
wcss = []
k_range = range(2, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range, wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Clusters')
plt.show()
```

```python
sil_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df)
    sil_scores.append(silhouette_score(df, labels))

plt.figure(figsize=(8,5))
plt.plot(k_range, sil_scores, marker='o', linestyle='--', color='red')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Analysis')
plt.show()
```

# **Apply** **K** **means** **clustering**

```python
k_optimal = 5  # Choose based on elbow and silhouette method
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df)
```

# **Visualisations**

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df.drop(columns=['Cluster']))
df_plot = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_plot['Cluster'] = df['Cluster']

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue=df_plot['Cluster'], palette='viridis', data=df_plot)
plt.title('Customer Segmentation using PCA')
plt.show()
```

```python
sns.pairplot(df, hue='Cluster', palette='coolwarm')
plt.show()
```

```python
fig = px.scatter_3d(
    df, x='Annual_Income_(k$)', y='Spending_Score', z='Age',
    color=df['Cluster'].astype(str), opacity=0.8,
    title='3D Customer Clusters'
)
fig.show()
```

```python
# Analyzing each cluster
for i in range(k_optimal):
    print(f"Cluster {i} Insights:")
    cluster_data = df[df['Cluster'] == i]
    print(cluster_data.describe())
    print("-" * 50)

# Defining business-oriented recommendations
def generate_recommendations(df):
    cluster_summary = df.groupby('Cluster').mean()

    recommendations = []

    for cluster, data in cluster_summary.iterrows():
        income = data['Annual_Income_(k$)']
        spending = data['Spending_Score']
        age = data['Age']

        if income > 0.6 and spending > 0.6:  # High-income, high spenders
            rec = f"Cluster {cluster}: *Premium customers* - Offer loyalty programs, VIP perks, and exclusive product recommendations."
        elif income < 0.4 and spending > 0.6:  # Low-income, high spenders
            rec = f"Cluster {cluster}: *Impulse buyers* - Promote discounts, installment payment options, and targeted ads."
        elif income > 0.6 and spending < 0.4:  # High-income, low spenders
            rec = f"Cluster {cluster}: *Potential big spenders* - Target with personalized promotions, luxury items, and premium services."
        elif age > 0.5 and spending < 0.4:  # Older, conservative spenders
            rec = f"Cluster {cluster}: *Cautious spenders* - Highlight value-for-money deals, loyalty points, and quality assurance."
        else:
            rec = f"Cluster {cluster}: *General customers* - Provide a balanced marketing strategy with personalized offers."

        recommendations.append(rec)

    return recommendations

# Generate and display recommendations
recommendations = generate_recommendations(df)
for rec in recommendations:
  print(rec)
```
