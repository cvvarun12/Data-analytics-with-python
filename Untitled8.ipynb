# **IMPORTING** **NECESSARY** **LIBRARIES**



```python
import pandas as pd

# Load the dataset
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_excel("Copy of Global_Superstore.xlsx")

# Display basic info and first few rows
df.info()
df.head()
```

## **DATA CLEANING**

```python
# Check for missing values
missing_values = df.isnull().sum()

# Handling missing values
df['Postal Code'].fillna(df['Postal Code'].median(), inplace=True)  # Filling missing postal codes with median
df.drop_duplicates(inplace=True)  # Remove duplicate rows

# Converting Date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Display cleaned data info
df.info()
```

# **DETECTING AND HANDLING OUTLIERS**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing outliers using boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Sales', 'Profit', 'Discount', 'Shipping Cost']])
plt.title("Boxplot of Sales, Profit, Discount, and Shipping Cost")
plt.show()

# Using IQR to remove outliers
Q1 = df[['Sales', 'Profit', 'Discount', 'Shipping Cost']].quantile(0.25)
Q3 = df[['Sales', 'Profit', 'Discount', 'Shipping Cost']].quantile(0.75)
IQR = Q3 - Q1

# Filtering out extreme outliers
df_cleaned = df[~((df[['Sales', 'Profit', 'Discount', 'Shipping Cost']] < (Q1 - 1.5 * IQR)) |
                   (df[['Sales', 'Profit', 'Discount', 'Shipping Cost']] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"Dataset size before outlier removal: {df.shape}")
print(f"Dataset size after outlier removal: {df_cleaned.shape}")
```

# **STATISTICAL** **ANALYSIS**

```python
# Descriptive statistics
statistics_summary = df_cleaned[['Sales', 'Profit', 'Discount', 'Shipping Cost']].describe()

# Correlation Matrix
correlation_matrix = df_cleaned[['Sales', 'Profit', 'Discount', 'Shipping Cost']].corr()

# Display summary statistics
print(statistics_summary)
print(correlation_matrix)
```

# **DATA** **VISUALISATION**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Sales'], bins=30, kde=True)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned['Profit'])
plt.title("Boxplot of Profit")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title("Correlation Heatmap")
plt.show()
```

# **SUMMARY** **REPORT**

```python
# Identifying trends and patterns
top_cities = df_cleaned.groupby('City')['Sales'].sum().nlargest(10)  # Top 10 cities by sales
top_products = df_cleaned.groupby('Product Name')['Sales'].sum().nlargest(10)  # Top 10 selling products
top_customers = df_cleaned.groupby('Customer Name')['Sales'].sum().nlargest(10)  # Top 10 customers

# Display insights
print("Top 10 Cities by Sales:\n", top_cities)
print("\nTop 10 Best-Selling Products:\n", top_products)
print("\nTop 10 Customers by Sales:\n",top_customers)
```

# **ADDITIONAL** **INSIGHTS** **AND** **VISUALISATIONS**

```python
# Aggregating sales and profit over time
df_cleaned['Year-Month'] = df_cleaned['Order Date'].dt.to_period('M')

# Monthly Sales and Profit Trends
plt.figure(figsize=(12, 6))
df_cleaned.groupby('Year-Month')[['Sales', 'Profit']].sum().plot(kind='line', marker='o', figsize=(12,6))
plt.title("Sales & Profit Trends Over Time")
plt.xlabel("Time (Year-Month)")
plt.ylabel("Total Sales & Profit")
plt.xticks(rotation=45)
plt.legend(['Sales', 'Profit'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
category_sales = df_cleaned.groupby('Category')['Sales'].sum().sort_values()
category_sales.plot(kind='barh', color='skyblue')
plt.title("Sales by Category")
plt.xlabel("Total Sales")
plt.ylabel("Category")
plt.show()

plt.figure(figsize=(14, 6))
subcategory_sales = df_cleaned.groupby('Sub-Category')['Sales'].sum().sort_values()
subcategory_sales.plot(kind='barh', color='lightcoral')
plt.title("Sales by Sub-Category")
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
region_profit = df_cleaned.groupby('Region')['Profit'].sum().sort_values()
region_profit.plot(kind='bar', color='seagreen')
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Ship Mode', y='Sales', data=df_cleaned, palette="coolwarm")
plt.title("Sales Distribution by Shipping Mode")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_cleaned['Discount'], y=df_cleaned['Profit'], alpha=0.5)
plt.title("Discount vs. Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Segment', y='Sales', data=df_cleaned, palette="viridis")
plt.title("Sales Distribution by Customer Segment")
plt.show()
```
